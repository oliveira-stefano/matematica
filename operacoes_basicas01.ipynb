# As 4 operações básicas da matemática

## Adição

### Propriedades da adição

- Propriedade comutativa da adição
    Alterar a soma das parcelas nao altera a soma. Por exemplo, $$ 4 + 2 = 2 + 4. $$
- Propriedade associativa da adição
    Alterar o agrupamento das parcelas nao altera a soma. Por exemplo, $$ (2+3) + 4 = 2 + (3+4) $$
- Propriedade do elemento neutro da adição
    A soma de 0 e qualquer numero e esse numero. Por exemplo, $$ 0 + 2 = 2 $$


```python
a = 2
b = 3
c = 4
```

```python
# propriedade comutativa
print(a + b)
print(b + a)
```

```python
# propriedade associativa
print((a + b) + c)
print(a + (b + c))
```

```python
# propriedade do elemento neutro
print(a + 0)
```

## Subtração

### Notas importantes acerca da subtração

- Oposto de um numero
    Por exemplo, -2 oposto 2 e -3 oposto 3.
- Sinal de negativo na frente de um parenteses
    $$ a - (b + c) = a - b - c $$
    $$ 8 + 3 - ( 4 - 9 ) = 8 + 3 - 4 + 9 = 11 - 4 + 9 = 16 $$

```python
# sinal de negativo na frente de um parenteses
print(a - (b + c))
```

## Multiplicação

*Regras de sinais*
*O produto entre dois numeros de sinais iguais eh positivo e o produto entre dois numeros de sinais diferentes eh negativo.*

```python
print(a * b)
print(a * (-b))
print(-a * b)
print((-a) * (-b))
```

*Notacao*
$$ 3 × 5 = 3 ⋅ 5 = 3(5) = (3)(5) = 15 $$

### Propriedades da multiplicação

- Propriedade do elemento neutro da multiplicacao
    O produto de 1 e qualquer numero eh esse numero. Por exemplo, $$ 1 * 2 = 2 $$
- Propriedade comutativa da multiplicacao
    A ordem dos fatores nao altera o produto. Por exemplo, $$ 2 * 3 = 3 * 2 = 6 $$
- Propriedade distributiva da multiplicacao
    Realizamos o produto do termo externo ao parenteses com os termos internos do parenteses. Por exemplo, $$ 1 * (2 + 3) = 1 * 2 + 1 * 3 = 2 + 3 = 5 $$
- Propriedade da anulacao
    O produto de 0 e qualquer numero eh 0. Por exemplo, 0 * 2 = 0.


## Divisão

*Regras de sinais*
A divisao entre dois numeros de sinais iguais eh positiva e a divisao entre dois numeros de sinais diferentes eh negativo.

*Notacao*
$$  3 ÷ 5 = 3 ⁄ 5 = \frac{3}{5} $$

Em uma operacao $$ 348 / 6 $$ 348 eh o dividendo(D) e 6 eh o (d)divisor 58 eh o (q)quociente e 0 eh o (r)resto.

$$ D = d * q + r $$

```python
q, r = divmod(a, b)
print("Quotient: ", q)
print("Remainder: ", r)
print("Formula D = d * q + r: ", b * q + r)
```
