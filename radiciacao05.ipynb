# Radiciação

## Definição

Radiciação é a operação inversa da potenciação

Seja $ a $ um número real e $ n $ um número natural diferente de zero. 
Dizemos que $ \sqrt[n]{a} $ é um numero $ b $, tal que $ ^{n}b = a$

## Exemplos

* a = $ \sqrt[3]{-8} = -2 $

$ \sqrt[n]{a} $ <br>
$n$ é o índice <br>
$\sqrt{}$ é o radical <br>
$a$ é o radicando <br>
